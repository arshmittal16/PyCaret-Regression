```
from pycaret.datasets import get_data
```

```
## Importing a list of datasets supported by PyCaret
dataSets = get_data('index')
dataSets
```

```
## Importing Boston DataSet - Regeression
boston_dataset = get_data('boston')
```

```
boston_dataset
```

```
## Removing Duplicates
```

```
boston_dataset.drop_duplicates()
```

```
## Assigning Dataset and Target Variables
from pycaret.regression import *
rege = setup(data=boston_dataset, target='medv')
```

```
## Comparing Various Models
compare_models()
```

```
## Plotting Residuals Graph on the best model
model = create_model('gbr')
plot_model(model, plot='residuals')
```

```
plot_model(model, plot='error')
```

```
plot_model(model, plot='learning')
```

```
## Comparing Models using Data Normalization
setup(data = boston_dataset, target = 'medv', normalize = True, normalize_method = 'zscore')
compare_models()
```

```
## Comparing Models using Feature Selection
setup(data = boston_dataset, target = 'medv', feature_selection = True, feature_selection_threshold = 0.8)
compare_models()
```

```
## ## Comparing Models using Outliners Removal
setup(data = boston_dataset, target = 'medv', remove_outliers = True, outliers_threshold = 0.05)
```

```
compare_models()
```

```
## Parameter Tuning
par_tune = tune_model(model, n_iter=20, optimize='mae')
```

```
## Saving the Model
save_model(par_tune, 'Boston_DataSet')
```
